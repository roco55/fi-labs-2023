```python
key_2 = "уф"
key_3 = "хук"
key_4 = "мару"
key_5 = "денди"
key_10 = "рубикпуджи"
key_11 = "йомайояйого"
key_12 = "капутфантану"
```

```python
def get_num(letter):
    i = 0
    while(letter != alphabet[i] and i<len(alphabet)-1):
        i+=1
    return i

def cypher(text,key):
    m = len(key)
    solution = ""
    for i in range (0,len(text)):
        solution+=(alphabet[(get_num(text[i]) + get_num(key[i%m]))%len(alphabet)])
    return solution

def decypher(text,key):
    m = len(key)
    solution = ""
    for i in range (0,len(text)):
        solution+=(alphabet[(get_num(text[i]) - get_num(key[i%m]))%len(alphabet)])
    return solution
```

Підрахунок N(Y), I(Y)

```python
def frequency(letter,text):
    freq = 0
    for i in range(0,len(text)) :
        if letter == text[i]:
            freq+=1
    return freq

def index(text):
    summ = 0
    for i in range(0, len(alphabet)):
        summ += (frequency(alphabet[i], text) * (frequency(alphabet[i], text)-1))
    index = 1/(len(text)*(len(text)-1))*summ

    return index
```

Дугий спосіб пошуку ключа

```python
def Counter(text,letter):
    count = 0 
    for i in text:
        if i == letter:
            count+=1
    return count

def M_i_g (text,letter):
    summ = 0
    for t in range(0,len(alphabet)):
        probability = p_govna[alphabet[t]]
        temp = (letter+t)%len(alphabet)
        frequency = Counter(text, alphabet[temp])
        summ += probability*frequency
    return summ

def find_key_2(text,r):
    key = ""
    ans =[]
    y = blocks(text,r)
    for i in range(0,r):
        for g in range(0,len(alphabet)):
            tmp = M_i_g(y[i], g)
            ans.append(tmp)
        key += alphabet[ans.index(max(ans))]
        ans.clear()
    return key

key = find_key_2(text_2,r)
print(key)

```
