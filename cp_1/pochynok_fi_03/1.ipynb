```python
from pathlib import Path
import math
f = open("Martin Eden (Jack London) translation.txt", "r", encoding="utf-8")
a = f.read()
a = a.lower()
b = a.replace(' ', '')
print(len(a))
print(len(b))
```

a - текст з пробілами

b - текст без пробілів

w - наявні пробіли

!w - пробіли відстуні

в кінці наведені обрахунки останніх двох пунктів

Calculating monograms in text w spaces and their frequency 

```python
all_freq_w = {}
 
for i in a:
    if i in all_freq_w:
        all_freq_w[i] += 1
    else:
        all_freq_w[i] = 1

    


myKeys_w = list(all_freq_w.keys())
myKeys_w.sort()
yesspaces = {i: all_freq_w[i] for i in myKeys_w}

#print(yesspaces)
yesspaces = {i:(all_freq_w[i]/len(a))for i in myKeys_w}
#print(yesspaces)

h1=0
for i in yesspaces.values():
    h1 = h1 + i*math.log2(i)
h1=h1*-1
print("H =", h1)

```

Calculating monograms in text !w spaces and their frequency 

```python
all_freq_l = {}
 
for i in b:
    if i in all_freq_l:
        all_freq_l[i] += 1
    else:
        all_freq_l[i] = 1   

myKeys_l = list(all_freq_l.keys())
myKeys_l.sort()
nospaces = {i: all_freq_l[i] for i in myKeys_l}

# print(nospaces) 

nospaces = {i:(all_freq_l[i]/len(b))for i in myKeys_l}
# print(nospaces)

h2=0
for i in nospaces.values():
    h2 = h2 + i*math.log2(i)
h2=h2*-1
print("H =", h2)

```

Calculating bigrams in text w spaces and their frequency [1,2] [2,3]

```python

d_all_freq_w = {}
prev_char = a[0]
for i in a:
    double = i + prev_char
    prev_char = i
    if double in d_all_freq_w:
        d_all_freq_w[double] += 1
    else:
        d_all_freq_w[double] = 1   

d_myKeys_w = list(d_all_freq_w.keys())
d_myKeys_w.sort()
dyesspaces = {i: d_all_freq_w[i] for i in d_myKeys_w}

# print(dyesspaces)

dyesspaces = {i:(d_all_freq_w[i]/(sum(dyesspaces.values())))for i in d_myKeys_w}

# print(dyesspaces)

h3=0
for i in dyesspaces.values():
    h3= h3 + i*math.log2(i)
h3=(h3*-1)/2
print("H =", h3)

```
