Написання допоміжних функцій


```python

def gcd(a, n):
    u = [1, 0]
    v = [0, 1]
    if(a ==0 or n ==0):
        return(0,[],[])
    if a > n:
        rn = a
        rn1 = n
        rn2 = 1
    elif a < n:
        rn = n
        rn1 = a
        rn2 = 1
        u,v = v, u
    else:
        rn2 = 0
        return(0,[],[])
    i = 0
    while rn2 != 0:
        i += 1
        q = int(rn / rn1)
        u.append(u[i-1]-u[i]*q)
        v.append(v[i-1]-v[i]*q)
        rn2 = rn - q * rn1
        #print(f"{rn} = {rn1}*{q} + {rn2}")
        rn = rn1
        rn1 = rn2
    return (abs(rn),u,v)



def clear(a):
    index = []
    for i in range(len(a)):
        if(a[i] == None):
            index.append(i)
    a = [a[i] for i in range(len(a)) if i not in index]
    return a


def inverse(a,n):
    u = gcd(a,n)[1]
    #print(u)
    invers = u[len(u)-2]
    if(invers<0):
        invers+=n
    return invers
inverse(15,31)

def linconj(a,n,b):
    solutions =[]
    d, u, v = gcd(a,n)
    if(d ==1):
        return([inverse(a,n)*b%n])
    if(d > 1):
        if(b%d != 0):
            return(solutions)
        else:
            #print(a,n,b,d)
            a1, n1, b1 = a/d, n/d , b/d
            u1 = gcd(a1,n1)[1]
            x0 = inverse(a1,n1)*b1%n1
            #print("x0 = ", x0, )
            for i in range(0,d):
                solutions.append(x0+i*n1)
            return(solutions)
```
